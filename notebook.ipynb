```python
#copy or assign the data to another variable to make the change comfortably.
r_l_c=r_level.copy()
r_r_c=r_rainfall.copy()
```

```python
#split the date column into month,day and year
r_l_c['Day']=[d.split('-')[0] for d in r_l_c.Date]
r_l_c['Month']=[d.split('-')[1] for d in r_l_c.Date]
r_l_c['Year']=[d.split('-')[2] for d in r_l_c.Date]
```

```python
print(r_l_c.head())#print the data head
```

```python
import seaborn as sb#import the library
```

```python
#This is the boxplot in which the range of rainfall level over the year is plotted for Poondi region.
base_color = sb.color_palette()[0]
# ax1 = sb.violinplot(data = r_l_c, x = 'Year', y = 'POONDI', color = base_color)
sb.boxplot(data = r_l_c, x = 'Year', y = 'POONDI', color = base_color)
```

```python
#This is the boxplot in which the range of rainfall level over the year is plotted for Cholavaram region.
sb.boxplot(data = r_l_c, x = 'Year', y = 'CHOLAVARAM', color = base_color)
```

```python
#This is the boxplot in which the range of rainfall level over the year is plotted for Redhills region.
sb.boxplot(data = r_l_c, x = 'Year', y = 'REDHILLS', color = base_color)
```

```python
#This is the boxplot in which the range of rainfall level over the year is plotted for Chembarambakkam region.
sb.boxplot(data = r_l_c, x = 'Year', y = 'CHEMBARAMBAKKAM', color = base_color)
```

From the above bar plot,We get rainfall level over the year.Different region have different rainfall level,but the rainfall level in 2019 is very low for all the reservoir region.CHEMBARAMBAKKAM has the lowest rainfall level in 2019.

```python
#This is the bar plot plotted to get the change in water level over the year
import matplotlib.pyplot as plt
r_l_c_year = r_l_c.groupby('Year')['POONDI'].sum()
r_l_c_year.plot(kind='bar');
plt.xticks(rotation=60);
plt.ylabel("rainfall level over year")
```

This plot show that poondi region has highest rainfall level in the year 2011 and least in the year 2004 and 2019.From the year 2013 the rainfall level decreasing,though the year 2016 have better rainfall level and again the level decrease.
